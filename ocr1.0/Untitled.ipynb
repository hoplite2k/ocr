```python
import pytesseract
import os
from pytesseract import Output
from PIL import Image  
from gtts import gTTS 
```

```python
def imageprocessing(image):
    results = pytesseract.image_to_data(image, output_type=Output.DICT, lang='kan')
    return results
```

```python
def textoutputs(results):
    file = ''
    for i in range(len(results["text"])):
        text = results["text"][i]
        file = file+text+" "
    return file
```

```python
def fileformation(text, filepath):
    text = text.strip()
    text = text.replace("  ","\n")
    
    with open(filepath, 'w') as file:
        file.write(text)    
```

```python
def txt2speech(filename, audiopath):
    with open(filename, 'r') as f:
        file_text = f.read()
    language = 'kn'
    audio = gTTS(text=file_text, lang=language, slow=False)
    audio.save(audiopath)
```

```python
path = './a.jpg'
filepath = './a.txt'
audiopath = "a.mp3"

image = Image.open(path)

results = imageprocessing(image)
text = textoutputs(results)
fileformation(text, filepath)

txt2speech(filepath, audiopath)
```
